```python
from collections import namedtuple

Pessoa = namedtuple('Pessoa', ['cpf', 'nome', 'idade', 'sexo', 'renda', "estado"])
```

```python
def cadastrar_usuario():
  continuar_cadastro = True
  cadastros = {}

  while continuar_cadastro:
    cadastrar = input('Deseja inserir um novo cadastro?')
    if cadastrar.lower() in ['sim', 's', 'yes', 'y']:
      cpf = input('Insira o cpf: ')
      nome = input('Insira o nome: ')
      idade = int(input('Insira a idade: '))
      sexo = input('Insira o sexo: ')
      renda = float(input('Insira a renda: '))
      estado = input('Insira o estado: ')
      cadastro = Pessoa(cpf=cpf, nome=nome, idade=idade, sexo=sexo, renda=renda, estado=estado)
      cadastros[cpf] = cadastro
    elif cadastrar.lower() in ['não', 'n', 'nao', 'no']:
      print(f'Cadastro finalizado, inserido {len(cadastros)} novos')
      continuar_cadastro = False
    else:
      print(f'Opção invalida. Foi inserido "{cadastrar}",\
              escolha entre [sim, s, yes, y] para cadastrar uma nova pessoa ou [não, nao, n, no] para parar')
  return cadastros
```

```python
def calcule_media_idade_por_sexo(cadastros):
  masc, soma_masc, media_masc = 0, 0, 0
  fem, soma_fem, media_fem = 0, 0, 0
  for cadastro in cadastros.values():
    if cadastro.sexo.lower() == 'm':
      masc += 1
      soma_masc += cadastro.idade
    elif cadastro.sexo.lower() == 'f':
      fem += 1
      soma_fem += cadastro.idade
    else:
      print('Sexo não identificado!')
  if masc > 0:
    media_masc = soma_masc / masc
  if fem > 0:
    media_fem = soma_fem / fem
  return (('M', media_masc), ('F', media_fem))
```

```python
def conte_quantidade_por_sexo(cadastros):
  masc, fem = 0, 0
  for cadastro in cadastros.values():
    if cadastro.sexo.lower() == 'm':
      masc += 1
    elif cadastro.sexo.lower() == 'f':
      fem += 1
    else:
      print('Sexo não identificado!')
  
  return (('M', masc), ('F', fem))
```

```python
def filtre_dados(cadastros): #def filtre_dados(cadastros, estados)
  estados = {}
  result = []
  if len(cadastros) == 0:
    print('Não existem pessoas cadastradas.')
  else:
    for cadastro in cadastros.values():
      estado = cadastro.estado
      if estado not in estados:
        estados[estado] = 1
      elif estado in estados:
        estados[str(estado)] += 1
  for estado in estados.items():
    result.append(estado)

  return tuple(result)
```

```python
def delete_cadastro(cadastros, cpf):
  del cadastros[str(cpf)]

  return cadastros
```

```python
# Chamando todas as funções e mostrando os resultados na tela
cadastros = cadastrar_usuario()
print(cadastros)
print(calcule_media_idade_por_sexo(cadastros))
print(conte_quantidade_por_sexo(cadastros))
print(filtre_dados(cadastros))
print(delete_cadastro(cadastros, int(input('Digite o número do CPF que deseja deletar: '))))
```
